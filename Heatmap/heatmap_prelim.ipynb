```python
import pandas as pd

lotto_data = pd.read_csv("C:/Users/denni/OneDrive/Desktop/Python Projects/NYLotto/winNumbers_fullDF.csv", )
lotto_data
```

```python
first_counts = pd.DataFrame({"Ball": pd.Series(lotto_data["1st Draw"]).value_counts().index,
                              "1st Draw": pd.Series(lotto_data["1st Draw"]).value_counts()})
second_counts = pd.DataFrame({"Ball": pd.Series(lotto_data["2nd Draw"]).value_counts().index,
                              "2nd Draw": pd.Series(lotto_data["2nd Draw"]).value_counts()})
third_counts = pd.DataFrame({"Ball": pd.Series(lotto_data["3rd Draw"]).value_counts().index,
                              "3rd Draw": pd.Series(lotto_data["3rd Draw"]).value_counts()})
fourth_counts = pd.DataFrame({"Ball": pd.Series(lotto_data["4th Draw"]).value_counts().index,
                              "4th Draw": pd.Series(lotto_data["4th Draw"]).value_counts()})
fifth_counts = pd.DataFrame({"Ball": pd.Series(lotto_data["5th Draw"]).value_counts().index,
                              "5th Draw": pd.Series(lotto_data["5th Draw"]).value_counts()})
sixth_counts = pd.DataFrame({"Ball": pd.Series(lotto_data["6th Draw"]).value_counts().index,
                              "6th Draw": pd.Series(lotto_data["6th Draw"]).value_counts()})
bonus_counts = pd.DataFrame({"Ball": pd.Series(lotto_data["Bonus #"]).value_counts().index,
                              "Bonus Draw": pd.Series(lotto_data["Bonus #"]).value_counts()})

counts_list = [first_counts, second_counts, third_counts, fourth_counts, fifth_counts, sixth_counts, bonus_counts]
```
