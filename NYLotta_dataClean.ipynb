```python
import numpy
import pandas

ny_lotto = pandas.read_csv("C:/Users/denni/Downloads/Lottery_NY_Lotto_Winning_Numbers__Beginning_2001.csv")

Dates = ny_lotto["Draw Date"].str.split(pat = "/", expand = True)
index_years = list(Dates[2])
index_months = list(Dates[0])
index_days = list(Dates[1])

hier_index = list(zip(index_years, index_months, index_days))
hier_index = pandas.MultiIndex.from_tuples(hier_index)
hier_index
```

```python
winNumber = ny_lotto["Winning Numbers"].str.split(expand = True)

winNumbers_fullDF = pandas.DataFrame(winNumber.to_numpy(), 
                                     hier_index, 
                                     ["1st Draw", "2nd Draw", "3rd Draw", "4th Draw", "5th Draw", "6th Draw"])
winNumbers_fullDF["Bonus #"] = list(ny_lotto["Bonus #"])
winNumbers_fullDF
```

```python
winNumbers_fullDF.index.names = ["Year", "Month", "Day"]
winNumbers_fullDF
```
